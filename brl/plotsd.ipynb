```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
r=np.arange(0,1,.01)
```

```python
y=20*r**2+r+90*r**3+50*np.sin(40*r)
```

```python
%config InlineBackend.figure_format = 'retina'
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
matplotlib.rcParams.update({'font.size': 50})

plt.scatter(r,y, linewidth=5)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('$y=20x^2+x+90x^3+50sin(40x)$')
plt.savefig('nonlinere.png')
```

```python
from scipy.spatial import distance_matrix
```

```python
x=np.arange(0,1,.01).reshape(-1,1)
sigma_v=1
l=0.3
```

```python
plt.imshow(sigma_v*np.exp(-1*distance_matrix(x,x)**2/(2*l**2)), origin='upper')
plt.colorbar()
```

```python
%config InlineBackend.figure_format = 'retina'
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
matplotlib.rcParams.update({'font.size': 50})
plt.imshow(sigma_v*np.exp(-1*distance_matrix(x,x)**2/(2*l**2)), origin='upper')
plt.savefig('rbf.png')
```
